```julia
using Plots
data = readdlm("data.txt")
x = data[:,1]
y = data[:,2]
plot(x,y,seriestype=:scatter,title="Data")
```

```julia
using Knet
predict(w,x) = w[1]*x .+ w[2]
J(w,x,y) = mean(abs2,y-predict(w,x))
dJ = grad(J)
w = Any[ 0.1*randn(1,1), 0.1*randn(1,1) ]
data = readdlm("data.txt")
x = data[:,1:1]'
y = data[:,2:2]'
for t=1:5
    println(J(w,x,y))
    w = w - 0.0001 * dJ(w,x,y)
end

```
